## Proses Crawling data Twitter

untuk data yang diambil dengan kata kunci **anies baswedan**

```python
twi = twint.Config()
# Proses Penyimpanan Data yang sudah di crawling
path = "/content/drive/MyDrive/prosaindata/DataCrawling/Data_twitter.csv"
twi.Search = 'anies baswedan'
# twi.Location = True
twi.Store_csv = True
twi.Custom["tweet"] = ["date","language", "username","name", "tweet"]
twi.Limit = 500
if( os.path.exists(path) == True):
    os.remove(path)

twi.Output = path
# try:
# twint.run.Search(twi)
twint.run.Search(twi)
# except:
#     time.sleep(6)
```
