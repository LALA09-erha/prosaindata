Data crawling adalah proses automatis untuk mengumpulkan dan mengindeks data dari berbagai sumber seperti situs web, database, atau dokumen. Proses ini menggunakan software atau aplikasi khusus yang disebut "twint" untuk mengakses sumber data dan mengambil informasi yang dibutuhkan.

## Inisialisasi Variable

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
```

## Proses Crawling Data Anies Baswedan

```python
query = "anies baswedan"
tweets = []
limit = 500

# tabel a id, sepal length postgre local
#  tabel b id , sepal width -"- cloud
# table c id, petal length mysql local
#table d id,petal width + class sql server local
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
      
        tweets.append([tweet.date, tweet.username, tweet.content])
        
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
print(df)

# to save to csv
path = "/content/drive/MyDrive/prosaindata/DataCrawling/Data_twitter_snt.csv"
if( os.path.exists(path) == True):
    os.remove(path)
df.to_csv(path)
```
