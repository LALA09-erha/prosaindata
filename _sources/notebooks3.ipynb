Data scraping atau yang juga sering disebut data extraction merupakan teknik atau metode otomatisasi yang memungkinkan seseorang untuk mengekstrak data dari sebuah website, database, aplikasi enterprise, atau sistem legacy yang kemudian dapat menyimpannya ke dalam sebuah file dengan format tabular atau spreadsheet.

Libary untuk scraping tersedia banyak, namun untuk kasus ini menggunakan libary `"BeautifulSoup"`

## Install Libary

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

## Inisialisasi Libary

```python
wiki_link = "https://www.tribunnews.com/"
html = urlopen(wiki_link).read()
soup = BeautifulSoup(html, 'html.parser')
```

## Proses Crawling data 

```python
# get sub menu
content_jurnal = soup.find("div", {"class": "main-menu"})
data_links = []
for i in content_jurnal.findAll("a"):
    if(i['href'] != "#"):
      data_links.append(i['href'])

print(data_links)
```

```python
# get artikel by menu
data_artikels = []

for i in data_links:
  menu_link = i
  html = urlopen(menu_link).read()
  soup = BeautifulSoup(html, 'html.parser')
  artikels = soup.find("ul", {"id": "latestul"})
  if(len(data_artikels) !=50):
    for artikel in artikels.findAll("a",class_="f20 ln24 fbo txt-oev-2"):
      if(len(data_artikels) != 50):
        data_artikels.append(artikel['href'])
      else:
        break
  else:
    break
print(data_artikels)
print(len(data_artikels))

```

```python
# mengecek apakah artikel berlanjut
data_page = []
ind = 0
for i in data_artikels:
    page_link = i
    page = urlopen(page_link).read()
    soup_page = BeautifulSoup(page, 'html.parser')
    page = soup_page.find("div", {"class": "pt10 pb5 plr10 ovh bggrey"})
    if(page == None):
      data_page.append([i])
      ind +=1
      # print(None)
    else:
      links = page.findAll("a")
      for link in links:
        if(i == link['href']):
          data_page.append([i])          
        else:
          data_page[ind].append(link['href'])
    
print(data_page)
print(len(data_page))
```

```python
data_csv = []
index = 0
for i in data_page:
  title_temp = ""
  artikel_temp = ""
  for page in i:
    artikel_link = page
    artikels_html = urlopen(artikel_link).read()
    soup_artikels = BeautifulSoup(artikels_html, 'html.parser')
    # get title
    title = soup_artikels.find("h1", {"id": "arttitle"})
    title_temp = title.text

    # get artikel
    artikel_hasil = soup_artikels.find("div", {"class": "side-article txt-article multi-fontsize"})
    for ark in artikel_hasil.findAll("p"):
      artikel_temp += ark.text+"\n"
  # print(artikel_temp)
  # print("--------------------------------")
  data_csv.append([title_temp,artikel_temp])
print((data_csv))
print(len(data_csv))
```

## Proses Penyimpanan data

```python
# list to dataframe
import pandas as pd
import os
path = "/content/drive/MyDrive/prosaindata/DataCrawling/Data_berita.csv"
df = pd.DataFrame(data_csv, columns =[ 'Judul','Berita'])

# save to csv
if( os.path.exists(path) == True):
    os.remove(path)
df.to_csv(path)
df
```
